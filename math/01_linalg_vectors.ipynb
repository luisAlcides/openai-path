## Tipos de normas

```python
# numpy
# vector
a = np.array([-2, 5, -1])

# normas
norma_l1 = np.linalg.norm(a, 1) # suma de valores absolutos
norma_l2 = np.linalg.norm(a, 2) # norma euclidiana
norma_linf = np.linalg.norm(a, np.inf) # Linf -> valor absoluto maximo

print(f"norma_l1: {norma_l1}")
print(f"norma_l2: {norma_l2}")
print(f"norma_linf: {norma_linf}")

```

```python
# math
a = [-2, 5, -1]

# l1: suma de valores absolutos
norma_L1 = sum(abs(x) for x in a)

# l2: raiz cuadrada de la suma de cuadrados
norma_L2 = math.sqrt(sum(x**2 for x in a))

# linf: maximo valor absoluto
norma_Linf = max(abs(x) for x in a)

print("L1:", norma_L1)     # 8.0
print("L2:", norma_L2)     # 5.477225575051661
print("L∞:", norma_Linf)   # 5.0
```

```python
import numpy as np
import matplotlib.pyplot as plt

# crear una malla de puntos en el plano
x = np.linspace(-3, 3, 400)
y = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x, y)


# calcular las normas
L1 = np.abs(X) + np.abs(Y)
L2 = np.sqrt(X**2 + Y**2)
Linf = np.maximum(np.abs(X), np.abs(Y))

# Dibujar las bolas unitas de cada norma
plt.figure(figsize=(8, 8))
plt.contour(X, Y, L1, levels=[1], colors='red', linewidths=2)
plt.contour(X, Y, L2, levels=[1], colors='blue', linewidths=2)
plt.contour(X, Y, Linf, levels=[1], colors='green', linewidths=2)

# Personalización
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.gca().set_aspect('equal', adjustable='box')
plt.title('Bolas unitarias para L1(rojo), L2(azul), Linf(verde)', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

u = np.array([2, 1], dtype=float)
v = np.array([1, 3], dtype=float)

# cálculos
dot_uv = float(np.dot(u, v))
norm_v = float(np.linalg.norm(v))
scalar_proj = dot_uv / norm_v # proyeccion escalar
vec_proj = (dot_uv / (norm_v**2)) * v # proyeccion vectorial

# preparar figura
plt.figure(figsize=(6, 6))

# vectores desde el origen
plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1)
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1)
plt.quiver(0, 0, vec_proj[0], vec_proj[1], angles='xy', scale_units='xy', scale=1)

# Linea perpendicular desde u hasta la recta de v
proj_point = vec_proj
plt.plot([u[0], proj_point[0]], [u[1], proj_point[1]], linestyle='--')

# configuracion de ejes
max_lim = max(3, np.max(np.abs([u, v])) + 0.5)
plt.xlim(-max_lim, max_lim)
plt.ylim(-max_lim, max_lim)
plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True, linestyle='--', alpha=0.5)
plt.title(f'Proyección de u sobre v\nu*v={dot_uv:.2f}, ||v||={norm_v:.2f}, proj_esc={scalar_proj:.3f}')
plt.show()

{"u": u.tolist(),
 "v": v.tolist(),
 "u·v": dot_uv,
 "||v||": norm_v,
 "proyección escalar (u sobre v)": scalar_proj,
 "proyección vectorial (u sobre v)": vec_proj.tolist()}
```

## Proyeccion vectorial
u = (0, 4), v = (1, 0)

```python
u = np.array([0 , 4])
v = np.array([1, 0])

producto = np.dot(u, v)
norma_v2 = np.linalg.norm(v)**2

proy_vec = (producto / norma_v2) * v
print(f"Proyeccion vectorial: {proy_vec}")
```

```python
u = (0, 4)
v = (1, 0)

producto = u[0]*v[0] + u[1]*v[1]
norma_v2 = v[0]**2 + v[1]**2 # norma al cuadrado

proy_vec = (producto / norma_v2) * v[0], (producto / norma_v2) * v[1]
print(f"Proyeccion vectorial: {proy_vec}")
```

# 📌 Vector Math Cheat Sheet

---

## 1. Vector Basics
Vector:  
$$u = (u_1, u_2, \dots, u_n)$$

**Addition:**  
$$u+v = (u_1+v_1, \; u_2+v_2, \dots)$$

**Subtraction:**  
$$u-v = (u_1-v_1, \; u_2-v_2, \dots)$$

---

## 2. Norms (Lengths)

**L2 norm (Euclidean):**  
$$\|u\|_2 = \sqrt{u_1^2 + u_2^2 + \dots + u_n^2}$$

**L1 norm (Manhattan):**  
$$\|u\|_1 = |u_1| + |u_2| + \dots + |u_n|$$

**Infinity norm:**  
$$\|u\|_\infty = \max(|u_1|, |u_2|, \dots, |u_n|)$$

---

## 3. Dot Product

$$u \cdot v = u_1 v_1 + u_2 v_2 + \dots + u_n v_n$$

**Geometric meaning:**  
$$u \cdot v = \|u\| \|v\| \cos(\theta)$$

- Positive → same direction  
- Zero → perpendicular  
- Negative → opposite  

---

## 4. Projection

**Scalar projection:**  
$$\text{proj}_v(u)_{\text{scalar}} = \frac{u \cdot v}{\|v\|}$$

**Vector projection:**  
$$\text{proj}_v(u) = \left(\frac{u \cdot v}{\|v\|^2}\right) v$$

---

## 5. Cross Product (3D)

$$u \times v = 
\begin{vmatrix}
\mathbf{i} & \mathbf{j} & \mathbf{k} \\
u_1 & u_2 & u_3 \\
v_1 & v_2 & v_3
\end{vmatrix}$$

- Result is perpendicular to both \(u\) and \(v\).  
- Magnitude = area of parallelogram.  

---

## 6. Angle Between Vectors

$$\cos(\theta) = \frac{u \cdot v}{\|u\| \|v\|}$$

## 3. Worker pulling a cart
u = (6, 8)
v = (0, 5)

```python
u = np.array([6,8])
v = np.array([0, 5])

uv = np.dot(u, v)
normv = np.linalg.norm(v)
scalar = uv / normv
vectorprojection = (uv / (normv)**2) * v

print("u*v = ", uv)
print("norm = ", normv)
print("scalar = ", scalar)
print("vector projection =", vectorprojection)
```
